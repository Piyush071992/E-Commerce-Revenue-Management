## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed.

### Ans 5

 The Total Revenue generated by the E-Commerce company over all orders placed is 2242221.750000008

```python
final_price=df['discounted_price']
sum1=0
sum2=0
sum3=0
sum4=0
for i in final_price:
    if i>600:
        x=i*0.25
        sum1+=x #sum1=sum1+x
        a=sum1
    elif i>350 and i<=600:
        y=i*0.15
        sum2+=y
        b=sum2
    elif i>100 and i<=350:
        z=i*0.1
        sum3+=z
        c=sum3
    else:
        w=i*0.05
        sum4+=w
        d=sum4


print('Total revenue of Ecommerce company=',a+b+c+sum4)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order


The top 10 brand with max. revenue in descending order is published with the help of groupby function. values are sorted in descending order for top 10 brands by using the head function and sum function is used to add it up

```python
x=[]
for i in df['discounted_price']:
  if i>600:
    x.append(i-i*0.25)
  elif i>350 and i<=600:
    x.append(i-i*0.15)
  elif i>100 and i<=350:
    x.append(i-i*0.1)
  else:
    x.append(i-i*0.05)

df['brand_rev']=x
df.groupby('Brand')['brand_rev'].sum().sort_values(ascending=False).head(10)
```
